```python
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
x=iris.data
y=iris.target
```

```python
#Comment or uncomment this cell to toggle standardisation
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit_transform(x)
```

<h2>5.1 a)</h2>
Training set = 75%, test set = 25%

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=69)
```

```python
model1=GaussianNB()
model2=DecisionTreeClassifier()
model3=KNeighborsClassifier()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

```python
y_pred1=model1.predict(x_test)
y_pred2=model2.predict(x_test)
y_pred3=model3.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Accuracy of:")
print("Naive Bayes : ",accuracy_score(y_pred=y_pred1,y_true=y_test))
print("Decision Tree : ",accuracy_score(y_pred=y_pred2,y_true=y_test))
print("K Nearest NeighbourW : ",accuracy_score(y_pred=y_pred2,y_true=y_test))
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
def score(model, x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    return accuracy_score(model.predict(x_test),y_test)
```

KNN Random Subsampling

```python
knn_acc=[]
for i in range(10):
    x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y)
    knn_acc.append(score(KNeighborsClassifier(), x_train,x_test,y_train,y_test))

print(sum(knn_acc)/10*100,"%")
```

KNN K Fold

```python
kf_knn_acc=[]
skf=StratifiedKFold(n_splits=5)
for train_index,test_index in skf.split(X=x,y=y):
    x_train,x_test,y_train,y_test=x[train_index],x[test_index],y[train_index],y[test_index]
    kf_knn_acc.append(score(KNeighborsClassifier(),x_train,x_test,y_train,y_test))
print(sum(kf_knn_acc)/5*100,"%")
```
